# FitMut2.0
In this notebook we walk through an example use case, generating data from simulation and then inferring fitness and establishment times. Though we expect most users will want to use our code to infer mutational parameters on their own experimental datasets, we include instructions on how to generate simulated data for the sake of completeness.

First we import that modules that we need to write the files that will be used as input parameters in our simulation.

```python
import numpy as np
import pandas as pd
import itertools
import csv
import os
```

```python
# input for fitsimu.py
import numpy as np
import pandas as pd
import itertools
import csv

dir_simulation = '../simulation/'

delta_t = 8
t_seq = np.arange(0, delta_t*15, delta_t)

tmp_1 = {'0':t_seq, 
         '1':10*np.ones(np.shape(t_seq)), 
         '2':20*np.ones(np.shape(t_seq)), 
         '3':50*np.ones(np.shape(t_seq)), 
         '4':100*np.ones(np.shape(t_seq))}

tmp = list(itertools.zip_longest(*list(tmp_1.values())))
with open(dir_simulation + 'simu_input_time_points.csv', 'w') as f:
    w = csv.writer(f)
    w.writerows(tmp)
        
        
step_size = 0.001
file_name_1_list = ['exponential_small', 'exponential_large', 'uniform_small', 'uniform_large']

for file_name_1 in file_name_1_list: 
    print(file_name_1)

    if file_name_1 == 'exponential_small':
        np.random.seed(5)
        s = np.random.exponential(0.045,100000) 
        bins_edge = np.arange(0, 0.145, step_size)
        
    elif file_name_1 == 'exponential_large':
        np.random.seed(5)
        s = np.random.exponential(0.075,100000)
        bins_edge = np.arange(0, 0.175, step_size)
        
    elif file_name_1 == 'uniform_small':
        np.random.seed(5)
        s = np.random.uniform(0,0.125,100000)
        bins_edge = np.arange(0, 0.125, step_size)
        
    elif file_name_1 == 'uniform_large':
        np.random.seed(5)
        s = np.random.uniform(0,0.16,100000)
        bins_edge = np.arange(0, 0.16, step_size)
        
    freq_bin = np.histogram(s, bins=bins_edge, density=True)[0] * step_size
    input_test = {'0':[1e-5], '1':bins_edge, '2':freq_bin}
    tmp = list(itertools.zip_longest(*list(input_test.values())))
    with open(dir_simulation + 'simu_input_%s_mutation_fitness.csv'%file_name_1, 'w') as f:
        w = csv.writer(f)
        w.writerows(tmp)
     
```

```python
directory_simulation = '../simulation/simulation_single_1e5_2022_06_05/'
directory_code = '../code/'

file_name_1_list = ['exponential_small', 'exponential_large', 'uniform_small', 'uniform_large']
file_name_2_list = ['0', '1', '2', '3']

plot_trajectory_noncolor = True
plot_trajectory_color = True


####################
csv_input = pd.read_csv(directory_simulation + 'simu_input_time_points.csv', low_memory=False, header=None)
t_seq = np.array(csv_input[0][~pd.isnull(csv_input[0])], dtype=float)
t_delta = 8
cell_num_average_bottleneck = 100

for file_name_1 in file_name_1_list: 
    print(file_name_1)
    
    csv_input = pd.read_csv(directory_simulation + 'simu_%s_0_EvoSimulation_Read_Number.csv'%file_name_1,
                            low_memory=False, header=None)
    lineages_num = csv_input.shape[0]
    
    cell_depth_seq = cell_num_average_bottleneck*lineages_num*np.ones(t_seq.shape)*t_delta
    input_tmp = {'0':t_seq, '1':cell_depth_seq}
    tmp = list(itertools.zip_longest(*list(input_tmp.values())))
    with open(directory_simulation + 'fitmut_input_%s_time_points.csv'%file_name_1, 'w') as f:
        w = csv.writer(f)
        w.writerows(tmp)
    
    ####################
    csv_input = pd.read_csv(directory_simulation + 'simu_' + file_name_1 +
                            '_EvoSimulation_Mutation_Info.csv', low_memory=False)    
    tmp_s = np.array(csv_input['mutation_fitness'], dtype = float)  #s
    tmp_t = np.array(csv_input['occurance_time'], dtype = float)  #occurrence time
    tmp_tau = np.array(csv_input['establishment_time'], dtype = int)  #establishment time
    tmp_mutant_num = np.array(csv_input['mutation_number'], dtype = int) #i-th mutation
    tmp_cell_num = np.array(csv_input['cell_number'], dtype = int) #number of mutant cells
    tmp_label = np.array(csv_input['lineage_label'], dtype = int)  #lineage label
    
    #label for established mutations
    tmp_idx = np.array([[tmp_label[i], i] for i in range(len(tmp_label)) if tmp_tau[i]>-16])
    #tmp_idx = np.array([[tmp_label[i], i] for i in range(len(tmp_label)) if
    #(tmp_tau[i]>-16)*(tmp_cell_num[i]>0)]) #label for established & non-extinct mutations
   
    idx_established_simulated = np.array(list(set(tmp_idx[:,0])))
    
    s_simulated = np.zeros(lineages_num)
    t_simulated = np.nan*np.zeros(lineages_num)
    tau_simulated = np.nan*np.zeros(lineages_num)
    s_simulated[tmp_idx[:,0]] = tmp_s[tmp_idx[:,1]]
    t_simulated[tmp_idx[:,0]] = tmp_t[tmp_idx[:,1]]
    tau_simulated[tmp_idx[:,0]] = tmp_tau[tmp_idx[:,1]]
    
    #keep single mutation in a lineage (with largest read number) and remove all others
    tmp_idx_mutiple = np.array([tmp_idx[idx2[idx3]][0]
                                for idx1 in idx_established_simulated 
                                if len(np.where(tmp_idx[:,0]==idx1)[0])>1 
                                for idx2 in [np.where(tmp_idx[:,0]==idx1)[0]] 
                                for mutant_num in [tmp_mutant_num[tmp_idx[idx2,1]]] 
                                for idx3 in [np.where(mutant_num == np.max(mutant_num))[0]]])
    if len(tmp_idx_mutiple)>0:
        s_simulated[tmp_idx_mutiple[:,0]] =  tmp_s[tmp_idx_mutiple[:,1]]
        t_simulated[tmp_idx_mutiple[:,0]] =  tmp_t[tmp_idx_mutiple[:,1]]
        tau_simulated[tmp_idx_mutiple[:,0]] =  tmp_tau[tmp_idx_mutiple[:,1]]
            
        
    ####################
    true_output = {'Mutation_Fitness': s_simulated, 
                   'Establishment_Time': tau_simulated, 
                   'Occurance_Time': t_simulated, 
                   'Index_Established':idx_established_simulated}
    
    tmp = list(itertools.zip_longest(*list(true_output.values())))
    with open(directory_simulation + 'mutation_true_information_%s.csv'%file_name_1, 'w') as f:
        w = csv.writer(f)
        w.writerow(true_output.keys())
        w.writerows(tmp)
       
    ####################
    if plot_trajectory_noncolor == True:
        for file_name_2 in file_name_2_list:
            os.system('python ' + directory_code + 'fig_trajectory_20220331.py -i ' 
                  + directory_simulation + 'simu_%s_%s_EvoSimulation_Read_Number.csv -t '%(file_name_1,file_name_2) 
                  + directory_simulation + 'fitmut_input_%s_time_points.csv -o'%file_name_1 
                  + directory_simulation + 'figure_lineage_%s_%s'%(file_name_1,file_name_2))
            
    if plot_trajectory_color == True:
        tmp = pd.DataFrame(s_simulated)
        tmp.to_csv(directory_simulation + 'lineage_trajectory_input_x_%s.csv'%file_name_1,index=False,header=False)
    
        for file_name_2 in file_name_2_list:
             os.system('python ' + directory_code + 'fig_trajectory_20220331.py -i '%(file_name_1,file_name_2) 
                   + directory_simulation + 'simu_%s_%s_EvoSimulation_Read_Number.csv -t ' 
                   + directory_simulation + 'fitmut_input_%s_time_points.csv -c '%file_name_1
                   + directory_simulation + 'lineage_trajectory_input_x_%s.csv -o '%file_name_1
                   + directory_simulation + 'figure_lineage_color_%s_%s'%(file_name_1,file_name_2))
        
```
